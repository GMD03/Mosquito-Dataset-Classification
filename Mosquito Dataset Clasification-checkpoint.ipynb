```python
This project aims to develop an accurate and EfficientNetV2B0-based Convolutional Neural Network (CNN)
that will detect three (3) classifications of mosquito species (I added additional information 
about the common disseases these species commonly transmit). 

The dataset used in this project is sourced from Kaggle: 
    Mosquito dataset for classification cnn
    (https://www.kaggle.com/datasets/masud1901/mosquito-dataset-for-classification-cnn/data)
which is a meticulously curated collection of high-quality images featuring three major mosquito species: 
    Culex, Anopheles, and Aedes. 
With 3,000 images in total, the dataset was structured to ensure a balanced representation of the three species, 
    each having 1,000 images.

In this project, the dataset was split into: 
    70% Training
    15% Validation
    15% Testing

After the training and fine-tuning, the model achieved a test accuracy of 100%.
```

```python
Import Required Libraries
    This part defines the is_image_okay function, which checks if the file at img_path 
        - is a valid, readable image.
        - Converts the image to RGB if possible.
        - Returns True if the image is good, and False if it’s corrupted, unreadable, or not an image.
```

```python
import os, shutil, random
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.applications import EfficientNetV2B0
import seaborn as sns
from PIL import Image
from PIL import UnidentifiedImageError

def is_image_okay(img_path):
    try:
        img = Image.open(img_path)
        img.verify()  # Checks if it's a readable image file
        img = Image.open(img_path)  # Re-open for reading after verify()
        img = img.convert("RGB")    # Ensures it's RGB for model
        return True
    except (UnidentifiedImageError, OSError, ValueError):
        return False
```

```python
Load Image Datasets and Explore Class Balance
```

```python
DATASET_PATH = 'mosquito_data/Mosquito_dataset'
print("Species folders:", os.listdir(DATASET_PATH))
for sp in os.listdir(DATASET_PATH):
    folder = os.path.join(DATASET_PATH, sp)
    print(f"{sp}: {len(os.listdir(folder))} files")
    for fname in os.listdir(folder):
        if not fname.lower().endswith(('jpg', 'jpeg', 'png')):
            os.remove(os.path.join(folder, fname))
```

```python
Remove Corrupted or Unreadable Images
```

```python
for sp in os.listdir(DATASET_PATH):
    folder = os.path.join(DATASET_PATH, sp)
    for fname in os.listdir(folder):
        img_path = os.path.join(folder, fname)
        if not is_image_okay(img_path):
            os.remove(img_path)
            print("Removed:", img_path)
```

```python
Split Dataset into Train, Val, Test (70/15/15)
```

```python
base_dir = 'mosquito_split'
train_dir = os.path.join(base_dir, 'train')
val_dir = os.path.join(base_dir, 'val')
test_dir = os.path.join(base_dir, 'test')

def split_data(SOURCE, TRAINING, VALIDATION, TESTING, split_ratio=(0.7, 0.15, 0.15), seed=42):
    random.seed(seed)
    for category in os.listdir(SOURCE):
        files = os.listdir(os.path.join(SOURCE, category))
        files = [f for f in files if f.lower().endswith(('jpg', 'jpeg', 'png'))]
        random.shuffle(files)
        n = len(files)
        n_train = int(n * split_ratio[0])
        n_val = int(n * split_ratio[1])
        splits = [TRAINING, VALIDATION, TESTING]
        indices = [(0, n_train), (n_train, n_train + n_val), (n_train + n_val, n)]
        for (split_folder, (start, end)) in zip(splits, indices):
            cat_folder = os.path.join(split_folder, category)
            os.makedirs(cat_folder, exist_ok=True)
            for fname in files[start:end]:
                shutil.copy(os.path.join(SOURCE, category, fname), os.path.join(cat_folder, fname))

if not os.path.exists(base_dir):
    os.makedirs(base_dir, exist_ok=True)
    split_data(DATASET_PATH, train_dir, val_dir, test_dir)
```

```python
Load Image Datasets and Explore Class Balance (Again)
```

```python
IMG_SIZE = 224
BATCH_SIZE = 32

train_ds = keras.utils.image_dataset_from_directory(
    train_dir, image_size=(IMG_SIZE, IMG_SIZE), batch_size=BATCH_SIZE, shuffle=True, seed=42
)
val_ds = keras.utils.image_dataset_from_directory(
    val_dir, image_size=(IMG_SIZE, IMG_SIZE), batch_size=BATCH_SIZE, shuffle=True, seed=42
)
test_ds = keras.utils.image_dataset_from_directory(
    test_dir, image_size=(IMG_SIZE, IMG_SIZE), batch_size=BATCH_SIZE, shuffle=False
)

class_names = train_ds.class_names
print("Class labels:", class_names)
```

```python
Mapping Mosquito Species to Diseases Transmitted
```

```python
species_to_disease = {
    'AEDES': 'Dengue',
    'ANOPHELES': 'Malaria',
    'CULEX': 'West Nile Virus'
}
```

```python
Visualize Example Images from Train Set
```

```python
plt.figure(figsize=(8, 8))
for images, labels in train_ds.take(1):
    for i in range(9):  # 9 images for a 3x3 grid
        plt.subplot(3, 3, i + 1)
        plt.imshow(images[i].numpy().astype("uint8"))
        species = class_names[labels[i]]
        disease = species_to_disease.get(species, "Unknown")
        plt.title(f"{species}\n({disease})", fontsize=10)
        plt.axis("off")
plt.tight_layout()
plt.show()
```

```python
Data Augmentation Pipeline
```

```python
data_augmentation = keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.15),
    layers.RandomZoom(0.2),
    layers.RandomContrast(0.2),
])
```

```python
Build EfficientNetV2B0-based Model with Transfer Learning
```

```python
base_model = EfficientNetV2B0(
    include_top=False,
    input_shape=(IMG_SIZE, IMG_SIZE, 3),
    weights="imagenet"
)
base_model.trainable = False

inputs = keras.Input(shape=(IMG_SIZE, IMG_SIZE, 3))
x = data_augmentation(inputs)
x = tf.keras.applications.efficientnet_v2.preprocess_input(x)
x = base_model(x, training=False)
x = layers.GlobalAveragePooling2D()(x)
x = layers.BatchNormalization()(x)
x = layers.Dropout(0.4)(x)
x = layers.Dense(128, activation='relu')(x)
x = layers.BatchNormalization()(x)
outputs = layers.Dense(len(class_names), activation='softmax')(x)
model = keras.Model(inputs, outputs)
model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=1e-4),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
model.summary()
```

```python
Training the Model with EarlyStopping and Checkpoints
```

```python
callbacks = [
    keras.callbacks.EarlyStopping(monitor='val_loss', patience=6, restore_best_weights=True),
    keras.callbacks.ModelCheckpoint('best_mosquito_model.keras', monitor='val_accuracy', save_best_only=True)
]
history = model.fit(
    train_ds.prefetch(tf.data.AUTOTUNE),
    validation_data=val_ds.prefetch(tf.data.AUTOTUNE),
    epochs=50,
    callbacks=callbacks
)
```

```python
Fine-Tune the Model
```

```python
base_model.trainable = True
model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=1e-5),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
finetune_history = model.fit(
    train_ds.prefetch(tf.data.AUTOTUNE),
    validation_data=val_ds.prefetch(tf.data.AUTOTUNE),
    epochs=20,
    callbacks=callbacks
)
```

```python
Plot Learning Curve (Training & Validation Accuracy and Loss)
```

```python
def plot_learning_curves(history, finetune_history=None):
    acc = history.history['accuracy']
    val_acc = history.history['val_accuracy']
    loss = history.history['loss']
    val_loss = history.history['val_loss']
    
    if finetune_history is not None:
        acc += finetune_history.history['accuracy']
        val_acc += finetune_history.history['val_accuracy']
        loss += finetune_history.history['loss']
        val_loss += finetune_history.history['val_loss']
    
    epochs_range = range(1, len(acc) + 1)
    
    plt.figure(figsize=(14, 5))
    # Accuracy
    plt.subplot(1, 2, 1)
    plt.plot(epochs_range, acc, label='Train Accuracy')
    plt.plot(epochs_range, val_acc, label='Val Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.title('Training and Validation Accuracy')
    plt.legend()
    
    # Loss
    plt.subplot(1, 2, 2)
    plt.plot(epochs_range, loss, label='Train Loss')
    plt.plot(epochs_range, val_loss, label='Val Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('Training and Validation Loss')
    plt.legend()
    
    plt.tight_layout()
    plt.show()

plot_learning_curves(history, finetune_history)
```

```python
Evaluate on Test Set
```

```python
model.load_weights('best_mosquito_model.keras')
test_loss, test_acc = model.evaluate(test_ds)
print(f"Test accuracy: {test_acc*100:.2f}%")
```

```python
Batch Prediction With Disease Display
```

```python
for images, labels in test_ds.take(1):
    preds = model.predict(images)
    for i in range(len(images)):
        pred_label = class_names[np.argmax(preds[i])]
        disease = species_to_disease.get(pred_label, "Unknown")
        print(f"Sample {i+1}: Predicted Species: {pred_label}, Disease: {disease}, Confidence: {np.max(preds[i])*100:.2f}%")
```

```python
Classification Report and Confusion Matrix
```

```python
from sklearn.metrics import confusion_matrix, classification_report

y_true, y_pred = [], []
for images, labels in test_ds:
    preds = model.predict(images)
    y_true.extend(labels.numpy())
    y_pred.extend(np.argmax(preds, axis=1))

print(classification_report(y_true, y_pred, target_names=class_names))

cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(6,6))
sns.heatmap(cm, annot=True, fmt='d', xticklabels=class_names, yticklabels=class_names, cmap="Blues")
plt.xlabel("Predicted"); plt.ylabel("True"); plt.title("Confusion Matrix")
plt.show()
```

```python
Predict a New Image
```

```python
from tensorflow.keras.utils import load_img, img_to_array

def predict_images_with_disease(img_paths):
    plt.figure(figsize=(10, 10))
    for i, img_path in enumerate(img_paths[:9]):
        plt.subplot(3, 3, i + 1)
        if not is_image_okay(img_path):
            plt.text(0.5, 0.5, "Invalid Image\nor Not a Mosquito",
                     fontsize=12, ha='center', va='center', color='red')
            plt.axis("off")
            continue
        img = load_img(img_path, target_size=(IMG_SIZE, IMG_SIZE))
        img_array = img_to_array(img)
        img_array = np.expand_dims(img_array, axis=0)
        img_array = tf.keras.applications.efficientnet_v2.preprocess_input(img_array)
        pred = model.predict(img_array, verbose=0)
        confidence = np.max(pred) * 100
        pred_label = class_names[np.argmax(pred)]
        disease = species_to_disease.get(pred_label, "Unknown")
        plt.imshow(img)
        if confidence < 60:
            plt.title(f"Low Confidence\n({confidence:.1f}%)", fontsize=10, color='red')
        else:
            plt.title(f"{pred_label}\n({disease})\n{confidence:.1f}%", fontsize=10)
        plt.axis("off")
    plt.tight_layout()
    plt.show()
```

```python
This part defines the prediction and visualization of the species and their associated disease for up to 
    9 user-specified images in a 3x3 grid. 
It also defines the guidelines of the displaying of the output. In simple words, for each image path:
    - If the image is invalid, it displays an error message in that grid spot.
    - If valid:
            * The system loads and preprocesses the image for the model.
            * The system uses the trained model to predict the mosquito species and the associated disease.
            * The system shows the image with its predicted species, disease, 
                and the model’s confidence as the subplot title.
    - If the model’s confidence is low, it marks the prediction in red as a warning.

Since the first image of the grid, which is clearly an LDR who is not part of the dataset, was still accepted 
    and was tried to predict on by the model is because it is a working image and not a corrupted image or image 
    with an invalid format. However, despite being accepted and labeled, the confidence level of the said image 
    is evidently low in comparison to images which is part of the dataset.

    Moreover,
        - If the image is valid, whether that of a mosquito or not a mosquito, the system will attemp to classify 
          the image, but will be given low confidence level to indicate that the image is irrelevant to the dataset.
```
